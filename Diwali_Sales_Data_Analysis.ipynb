# Diwali Sales Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.describe()
```

```python
df.info()
```

```python
#Remove column Status and Unnamed1
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
#Finding null values in columns
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#Drop Null Values
df.dropna(inplace = True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
#Change of a Column Datatype
df['Amount']= df['Amount'].astype('int')
```

```python
df.columns
```

```python
#Rename of a Column
df.rename(columns={'Age Group': 'Age_Group'},inplace = True)
```

# EXPLORATORY DATA ANALYSIS

```python
df.info()
```

```python
df.columns
```

```python
#Visualize Gender column in chart
sns.countplot(x='Gender',data = df)
```

```python
#Add Data Values in the chart
ax = sns.countplot(x='Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group by Gender with Amount Column
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Gender', y = 'Amount',data = sales_gen,errwidth =0).set(title = "Gender vs Amount")
```

### From the above graph we can easily find that most buyers are Females compare than Males

# AGE

```python
ax = sns.countplot(data = df, x = 'Age_Group', hue = 'Gender').set(title = 'Age_Group VS Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Age = df.groupby(['Age_Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x = 'Age_Group', y = 'Amount',data = Sales_Age).set(title = 'Age_Group VS Amount')
```

### From the above graph we can see that most of buyers are between age group of 26 to 35 yrs Females

## STATE

```python
#Total no of orders from top 10 States
df.info()
```

```python
sales_States = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x = 'State', y = 'Orders',data = sales_States).set(title = "States VS Orders")
```

```python
sales_Amt = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x = 'State', y = 'Amount',data = sales_Amt).set(title = "States VS Amount")
```

### From the above graph we can see that most of the orders and total sales amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## MARITAL STATUS

```python
ax = sns.countplot(x='Marital_Status',data = df)

sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Marital_Status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount',data = sales_Marital_Status, hue = 'Gender').set(title = "Marital_Status VS Amount")
```

### From the above graph we can see that most of the buyers are married (Women) and they have the more purchasing power

```python
sns.set(rc={'figure.figsize': (25,10)})
ax = sns.countplot( x = 'Occupation' ,data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Amt = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(15)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x = 'Occupation', y = 'Amount',data = sales_Amt).set(title = "Occupation VS Amount")
```

### From the above graph we can see that most buyers are from IT, Healthcare and Aviation Sector

```python
sns.set(rc={'figure.figsize': (25,7)})
ax = sns.countplot( x = 'Product_Category' ,data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Product_catg = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x = 'Product_Category', y = 'Amount',data = sales_Product_catg).set(title = "Product_Category VS Amount")
```

### From the above graph we can see that most of the sold products are from Food, Clothing and Electronics Category

```python
sales_States = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x = 'Product_ID', y = 'Orders',data = sales_States).set(title = "Product_ID VS Orders")
```

```python
fig1, ax1= plt.subplots(figsize=(13,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind='bar').set(title = 'Product_ID VS Orders')
```

### Married age group between 26 to 35 from UP , Maharstra and Karnataka working in IT , Aviation and Healthcare sector are more likely to buy products from Food, Clothing and Electronics Category
